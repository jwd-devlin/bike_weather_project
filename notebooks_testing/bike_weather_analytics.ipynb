```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))

ax1 = df[["weather_date", "avg_trip_duration"]].plot(color='blue', grid=True)
ax2 = df[["weather_date", "avg_temperature"]].plot(color='red', grid=True, secondary_y=True)

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()


plt.legend(h1+h2, l1+l2, loc=2)
plt.show()
```
