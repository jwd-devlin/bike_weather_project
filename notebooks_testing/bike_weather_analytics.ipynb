```python
df = DumpMongoCollection("local").extract_pandas("average_daily_trip_duration_plus_count")
```

```python
df = df.dropna(subset = ["avg_temperature"], axis = 0).sort_values('weather_date')
```

```python
df.plot.scatter(x='avg_trip_duration',
                      y='avg_temperature',
                      c='DarkBlue')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(df["weather_date"], df["avg_trip_duration"])
ax1.set_ylabel('avg_trip_duration')
ax1.set_xticklabels(df["weather_date"], rotation=90)
ax2 = ax1.twinx()
ax2.plot(df["weather_date"], df["avg_temperature"], 'r-')
ax2.set_ylabel('avg_temperature', color='r')
```
